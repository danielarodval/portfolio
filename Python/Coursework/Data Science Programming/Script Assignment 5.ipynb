Use the "Stock_px" dataset and address the following functions:

```python
import pandas as pd
df = pd.read_csv('datasets/examples/stock_px.csv', parse_dates=True, index_col=0)
display(df)
```

## 1. Draw line plot for each of the stocks separately/individually - 2 points

```python
import matplotlib.pyplot as plt
for c in df.columns:
    df[c].plot(title=c)
    plt.show()
```

## 2. Draw line, scatter, box plot for each of the stocks within the same subplot - 2 points

```python
fig, axes = plt.subplots(3, 1, figsize=(10, 12))

#line plot
for c in df.columns:
    df[c].plot(ax=axes[0], title='Line Plot')
axes[0].legend()

#scatter plot
for c in df.columns:
    axes[1].scatter(df.index,df[c], label=c)
axes[1].set_title('Scatter Plot')

#box plot
df.boxplot(ax=axes[2])
axes[2].set_title('Box Plot')

plt.tight_layout()
plt.show()
```

## Draw a linear regression model fit for apple stock - 2 points

```python
import seaborn as sns
sns.regplot(x=df.index.astype(int), y=df['AAPL'])
plt.xlabel('Date')
plt.ylabel('Stock Price')
plt.show()
```

## 4. Draw a dotted line plot for SPX and mark all the peak points at various timesteps - 2 points

```python
from scipy.signal import find_peaks
peaks, _ = find_peaks(df['SPX'].values)

plt.figure(figsize=(10,6))
plt.plot(df.index, df['SPX'], 'b:', label='SPX')
plt.plot(df.index[peaks], df['SPX'].values[peaks], 'ro', label='SPX Peaks')
plt.legend()
plt.show()
```

## 5. Using Seaborn plot pairwise.  - 2 points

```python
sns.pairplot(df)
plt.suptitle('Pairwise Plots', y=1)
plt.show()
```
