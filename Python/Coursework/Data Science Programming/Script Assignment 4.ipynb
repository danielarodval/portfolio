## 1. Use unstack on years 1959-1999 and then stack for the remaining years in the years column. - 2 points

```python
import pandas as pd
df = pd.read_csv('datasets/examples/macrodata.csv')

df1 = df[df['year'].between(1959, 1999)]
df2 = df[~df['year'].between(1959, 1999)]

df1 = df1.set_index(['year']).unstack().reset_index()
df2 = df2.set_index(['year']).stack().reset_index()
display(df1)
display(df2)
```

## 2. Use Swaplevel on columns "cpi" and "realint" - 1 Point

## 3. Merge all columns that start with "real"  into new column "unreal". (Use RE from the previous notebook if needed) - 1 Point

```python
real_cols = df.filter(regex='^real').columns
df[real_cols].apply(lambda row: '_'.join(row.values.astype(str)), axis=1)
```

## 4. Provide an example for join() - 1 Point

```python
df1.set_index(['year', 'level_0'], inplace=True)
df2.set_index(['year', 'level_1'], inplace=True)
df1.join(df2, how='outer', lsuffix='_df1', rsuffix='_df2')
```

## 5. Provide an example for concat() with index enabled - 1 Point

```python
#need to redo the original dataframes
df1 = df[df['year'].between(1959, 1999)]
df2 = df[~df['year'].between(1959, 1999)]

df1 = df1.set_index(['year']).unstack().reset_index()
df2 = df2.set_index(['year']).stack().reset_index()

pd.concat([df1, df2], axis=1)
```
