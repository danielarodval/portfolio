Assignment 4: Fibonacci Sequence with Function

```julia
function fibonacci(n)
    if n == 1 || n == 2
        return 1
    else
        return fibonacci(n - 1) + fibonacci(n - 2)
    end
end
n = 6
result = fibonacci(n)
println("The $n-th Fibonacci number is: $result")
```

Assignment 5: Data Visualization

```julia
using Plots

x_values = -5:5
y_values = x_values .^ 2


plt = plot(x_values, y_values, label="y = x^2", xlabel="x", ylabel="y", linewidth=2)
title!("Line Graph of y = x^2")
display(plt)
```
