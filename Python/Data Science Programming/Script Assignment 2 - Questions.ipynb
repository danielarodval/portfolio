### 2. Reading tabular data (Use: read_csv) - 1 point

```python
df = pd.read_csv('educ_figdp_1_Data.csv')
df
```

### 3. Viewing Data (Use: head, tail, column, index, values, describe) - 1 point

```python
display(df.head())
display(df.tail())
display(df.columns)
display(df.index)
display(df.values)
display(df.describe())
```

### 4. Selection(Use: Value, select a subset of columns and rows (hint: slice)) - 1 point

```python
df[['TIME','GEO']].iloc[5:8]
```

### 5. Filtering Data (Use: any condition (for ex: filter those values less than or equal to 6.5)) - 1 point

```python
df[df['Value'] == ':'] = 0
df['Value'] = df['Value'].astype(float)
#df.dtypes
df[df['Value'] <= 6.5]
```

### 6. Filtering Missing Values (Use: Nan) - 1 point

```python
df[df.isna().any(axis=1)]
```

### 7. Manipulating Data (Use: drop, append, dropna, fillna) - 1 point

```python
display(df.drop(['INDIC_ED','Flag and Footnotes'], axis=1))
df2 = pd.DataFrame(index=range(0,351),columns=['A'], dtype='float')
display(df.append(df2))
display(df.dropna())
display(df.fillna(-1))
```

### 8. Sorting (Use: Sort_values, sort_index) - 1 point

```python
df.sort_values(by='Value')
```

### 9. Grouping Data (Use: Groupby) - 1 point

```python
df[['GEO','Value']].groupby('GEO').sum(numeric_only=True)
```

### 10. Rearrange Data(Use: pivot_table, rank) - 1 point

```python
table = pd.pivot_table(df, values='Value', index=['TIME'], columns=['GEO'], aggfunc="sum")
display(table)
display(df.rank)
```

### 11. Visualization the data (Use: Stacked bar chart or any of your choice) - Bonus - 5 point

```python
from matplotlib import pyplot as plt
plot = df[['GEO','Value']].groupby('GEO').sum(numeric_only=True)
plot = plot.iloc[1:]
plt.figure(figsize=(10, 6))  # Adjust the figure size as needed
plt.bar(x=plot.index, height=plot['Value'])
plt.xlabel('GEO')
plt.ylabel('Value')
plt.xticks(rotation=90)  # Rotate x-axis tick labels for better visibility
plt.tight_layout()  # Adjust layout to prevent clipping of labels
plt.show()
```
