### Data Acquisition

```python
import requests
response = requests.get("https://www.whitehouse.gov/state-of-the-union-2023/")
```

### Data Cleaning/Processing

```python
from bs4 import BeautifulSoup
page = BeautifulSoup(response.content, 'html.parser')
page_content_div = page.find_all("div",class_= "sotu-annotations__content")
paragraphs = page_content_div[0].find_all("p")
sou = ""
for paragraph in paragraphs:
    sou = sou + " " +paragraph.get_text()

import nltk
nltk.download('punkt')
from nltk.tokenize import word_tokenize
word_tokenize(sou)
```

```python
import numpy as np
#nltk.download('stopwords')
from nltk.corpus import stopwords
from PIL import Image
mask = np.array(Image.open("biden.jpg"))

from wordcloud import WordCloud
wordcloud = WordCloud(mask = mask, stopwords = stopwords.words('english'), background_color = "white").generate(sou)
```

### Data Visualization

```python
import plotly.express as px

fig = px.imshow(wordcloud)
fig.update_xaxes(showticklabels = False)
fig.update_yaxes(showticklabels = False)
```
